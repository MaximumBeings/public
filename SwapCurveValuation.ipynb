```python
ycHandle = ql.YieldTermStructureHandle(depoFutSwapTermStructure)
libor6m = ql.USDLibor(ql.Period(6,ql.Months), ycHandle)
type_ = ql.VanillaSwap.Receiver
cal = ql.TARGET()
fixedBdc = ql.ModifiedFollowing
floatBdc = libor6m.businessDayConvention()

fixedDc = ql.Thirty360(ql.Thirty360.BondBasis)
floatDc = libor6m.dayCounter()

fixing = ycHandle.referenceDate()
ql.Settings.instance().evaluationDate = cal.advance(fixing, -2,ql.Days)
mat = fixing + ql.Period(3,ql.Years)

fixedSched = ql.Schedule(fixing, mat, ql.Period(ql.Semiannual), cal, fixedBdc, \
                         fixedBdc, ql.DateGeneration.Backward, False)

floatSched = ql.Schedule(fixing, mat, ql.Period(ql.Semiannual), cal, floatBdc, \
                         floatBdc, ql.DateGeneration.Backward, False)

nominal = 10000000
spread = 0.0
fixedRate = 0.04

swap = ql.VanillaSwap(type_,nominal, fixedSched, fixedRate,
                     fixedDc, floatSched,libor6m,spread,floatDc)

swap.setPricingEngine(ql.DiscountingSwapEngine(ycHandle))

print ''
print "Fair Rate: ", swap.fairRate()
print ''
print "Fair Spread: " , swap.fairSpread()
print ''
print "NPV Unfair: " , swap.NPV()

fairRate = swap.fairRate()
fairSpread = swap.fairSpread()
bps = 0.0001

swapFair = ql.VanillaSwap(type_, nominal, fixedSched, fairRate, \
                          fixedDc, floatSched, libor6m,spread, floatDc)

swapFair.setPricingEngine(ql.DiscountingSwapEngine(ycHandle))
print ''
print "NPV fair: ", round(swapFair.NPV(),8)


swapBps = ql.VanillaSwap(type_, nominal, fixedSched, fairRate+bps, \
                          fixedDc, floatSched, libor6m,spread, floatDc)

swapBps.setPricingEngine(ql.DiscountingSwapEngine(ycHandle))

swapBpsNpvFix=swapBps.fixedLegNPV()

swapFairNpvFix=swapFair.fixedLegNPV()

npvDiffFix=swapBpsNpvFix-swapFairNpvFix

fixBps=swapFair.fixedLegBPS()

print ''
print "NPV Diff Fix: " , npvDiffFix
print ''
print "Fix Bps: ", fixBps
```
